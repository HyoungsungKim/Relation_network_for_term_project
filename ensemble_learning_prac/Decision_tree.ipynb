```python
from sklearn import tree
```

```python
X = [[0,0], [1,1]]
Y = [0, 1]
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,Y)
```

```python
clf.predict([[2., 2.]])
```

```python
clf.predict_proba([[2., 2.]])
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
X, y = load_iris(return_X_y=True)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

```python
tree.plot_tree(clf)
```
